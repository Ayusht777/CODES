## Float Formatting 
 


```python
# "{value:width/whitespaces.precision f}"
```

basically it allows to you to adjust the width and precision of a folating (folat) point number

```python
price_of_potatoes = 50/3
print(price_of_potatoes) 
```

```python
"the price of potatoes per kg {p:1.2f}".format(p = price_of_potatoes)
```

```python
note: percision here will count no after decimal point
    {v:10.5f}
    here it will show output
    25.12345
    {v:10.3f}
    25.123
    
```

```python
print("the price of potatoes per kg {p:11.5f}".format(p = price_of_potatoes))
```

# Alignment, padding and precision with `.format()`

String formatting with padding and alignment


Number --->>> String

```python
"checking something {l:<5}".format(l = "left")  
```

```python
"checking something {r:>10}".format(r = "right")
```

```python
"checking something {c:^12}".format(c = "center")
```

```python
# adding any character with padding
```

```python
"checking something {p:=<12}".format(p= "padding")    # with left alignment
```

```python
"checking something {p:->12}".format(p= "padding")    # with right alignment
```

```python
"checking something {p:#^6}".format(p= "pa")    # with center alignment
```
